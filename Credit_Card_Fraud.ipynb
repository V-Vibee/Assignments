````python
# Importing the Dependencies

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
````

````python
# Loading the Data

credit_card_data= pd.read_csv('/creditcard.csv')
````

````python
# First 5 rows in dataset

credit_card_data.head(5)
````

````python
# Last 5  rows

credit_card_data.tail(5)
````

````python
# Getting Dataset Informations

credit_card_data.info()
````

````python
credit_card_data.isnull().sum()
````

````python
# Data Distribution

credit_card_data['Class'].value_counts()
````

````python
# 1-----> Fraud Transaction & 0------> Normal transaction, implies imbalanced Dataset
````

````python
# Saving Legit trans separately

legit= credit_card_data[credit_card_data.Class== 0]
fraud= credit_card_data[credit_card_data.Class== 1]
````

````python
legit.shape
````

````python
fraud.shape
````

````python
# Statistical Measures

legit.Amount.describe()
````

````python
fraud.Amount.describe()
````

````python
# Compare the values for both Transaction

credit_card_data.groupby('Class').mean()
````

````python
# Dealing with imbalanced data, using Undersampling

# Build a sample dataset containing similar distribtion of normal transactions and the fraudulent transaction

# Fraud Transaction---> 492
````

````python
legit_sample= legit.sample(n=492)
````

Concatinating two dataframes
axis 0= Rows
axis 1= Columns

````python
new_dataset= pd.concat([legit_sample, fraud], axis=0)
````

````python
new_dataset.head()
````

````python
new_dataset.tail()
````

````python
new_dataset['Class'].value_counts()
````

````python
new_dataset.groupby('Class').mean()
````

# Splitting Dataset into features nd targets


```



````python
X= new_dataset.drop(columns='Class', axis=1)
Y=new_dataset['Class']
````

````python
print(X)
````

````python
print(Y)
````

````python
# Splitting data into Train/Test data


X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size= 0.2, stratify=Y, random_state=2)
````

````python
print(X.shape, X_train.shape,X_test.shape)
````

Model Training

````python
# Logistic Regression

model= LogisticRegression()
````

````python
# Training the Logistic Regression model with Training Data


model.fit(X_train, Y_train)
````

````python
# Evaluation of model/Performance

````

````python
#Accuracy Score on Training Data


X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)

````

````python
print('Accuracy on Training data :', training_data_accuracy)
````

````python
# Accuracy on Test Data


X_test_prediction = model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction, Y_test)
````

````python
print('Accuracy score on Test Data :', test_data_accuracy)
````
